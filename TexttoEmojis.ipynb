```python
import re
import nltk
from nltk.corpus import wordnet
from transformers import pipeline

nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
emoji_dict = {
    "love": "❤️",
    "pizza": "🍕",
    "happy": "😊",
    "sad": "😢",
    "dog": "🐶",
    "cat": "🐱",
    "fire": "🔥",
    "food": "🍽️",
    "coffee": "☕",
    "book": "📚",
    "music": "🎵"
}
```

```python
def get_synonyms(word):
    synonyms = set()
    for syn in wordnet.synsets(word):
        for lemma in syn.lemmas():
            synonyms.add(lemma.name().lower())
    return synonyms

expanded_dict = {}

for key, emoji in emoji_dict.items():
    expanded_dict[key] = emoji
    for syn in get_synonyms(key):
        expanded_dict[syn] = emoji

expanded_dict
```

```python
def text_to_emoji_basic(text):
    words = re.findall(r'\w+', text.lower())
    output = ""
    for w in words:
        output += expanded_dict.get(w, w) + " "
    return output.strip()
```

```python

emotion_model = pipeline("text-classification",
                         model="j-hartmann/emotion-english-distilroberta-base",
                         return_all_scores=False)

emotion_to_emoji = {
    "joy": "😊",
    "love": "❤️",
    "anger": "😡",
    "sadness": "😢",
    "fear": "😨",
    "surprise": "😲",
    "disgust": "🤢"
}

def text_to_emoji_ai(text):
    try:
        emotion = emotion_model(text)[0]['label'].lower()
        emo = emotion_to_emoji.get(emotion, "")
        return emo
    except:
        return ""
```

```python
def convert_text_to_emoji(text):
    rule_output = text_to_emoji_basic(text)
    emotion_emoji = text_to_emoji_ai(text)
    return rule_output + " " + emotion_emoji
```

```python
texts = [
    "I love pizza",
    "I am very sad today",
    "My dog is happy",
    "I am angry about this",
    "I want coffee",
    "I like music and books"
]

for t in texts:
    print(f"{t}  =  {convert_text_to_emoji(t)}")
```
